```python
import pandas as pd
```

### Filling the null values of datasets before importing into database

#### CovidDeaths CSV

```python
df = pd.read_csv("CovidDeaths.csv", parse_dates= True)
df.head(5)
```

```python
df.fillna("\\N", inplace= True)
df.head(5)
```

```python
df.to_csv("covid_deaths.csv", index= False, encoding= 'utf-8')
```

#### CovidVaccinations CSV

```python
df = pd.read_csv("CovidVaccinations.csv", parse_dates= True)
df.head(5)
```

```python
df.fillna("\\N", inplace= True)
df.head(5)
```

```python
df.to_csv("covid_vaccinations.csv", index= False, encoding= 'utf-8')
```
